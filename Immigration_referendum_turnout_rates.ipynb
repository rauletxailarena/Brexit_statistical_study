```python
import pymongo
import datetime
import collections

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import pandas as pd
import scipy.stats
from collections import Counter
from sklearn import cluster
```

```python
merged_df = pd.read_csv("data/anotated_immigration_net_referendum_df.csv")
```

```python
cluster0_df = merged_df[merged_df["Cluster"] == 0]
cluster1_df = merged_df[merged_df["Cluster"] == 1]
```

```python
b= cluster0_df.pivot_table(index="Region", aggfunc = np.mean) [ "Pct_Turnout"]
b.plot.bar()
```

```python
a= cluster1_df.pivot_table(index="Region", aggfunc = np.mean) [ "Pct_Turnout"]
a.plot.bar()
```

```python
plt.figure()
a.plot.bar(color= "red",figsize=(10,8), legend=False)
b.plot.bar(color= "black", alpha=0.4,legend=False)
plt.title("% turnout per region")
plt.legend(("Under 500,000 immigrants","Over 500,000 immigrants"), loc="center left", bbox_to_anchor=(1.0, 0.5))
```

```python
merged_turnout_df = pd.merge(pd.DataFrame(a).reset_index(), pd.DataFrame(b).reset_index(), on="Region")
merged_turnout_df.columns = [["Region", "Pct turnout under 20,000", "Pct turnout over 20,000"]]
merged_turnout_df
```
